Question 1: What is hypothesis testing in statistics?

Answer: Hypothesis testing is a statistical method used to determine if there is enough evidence in a sample data to infer that a certain condition is true for the entire population. It involves formulating a null hypothesis (H0) and an alternative hypothesis (H1), collecting data, and using statistical tests to decide whether to reject the null hypothesis in favor of the alternative hypothesis.

Question 2: What is the null hypothesis, and how does it differ from the alternative hypothesis?

Answer: The null hypothesis (H0) is a statement of no effect, no difference, or no relationship. It represents the status quo or a commonly held belief. The alternative hypothesis (H1), on the other hand, is a statement that contradicts the null hypothesis and suggests that there is a significant effect, difference, or relationship. The goal of hypothesis testing is to determine if there is enough evidence to reject the null hypothesis in favor of the alternative hypothesis.

Question 3: Explain the significance level in hypothesis testing and its role in deciding the outcome of a test.

Answer: The significance level (alpha or α) is a threshold used in hypothesis testing to determine whether to reject the null hypothesis. It represents the probability of rejecting the null hypothesis when it is actually true (Type I error). A common significance level is 0.05 (or 5%), which means there is a 5% chance of incorrectly rejecting the null hypothesis. If the p-value (the probability of observing the data if the null hypothesis is true) is less than or equal to the significance level, the null hypothesis is rejected.

Question 4: What are Type I and Type II errors? Give examples of each.

Answer: Type I and Type II errors are potential mistakes that can occur during hypothesis testing.

*   **Type I Error (False Positive):** This happens when you reject the null hypothesis when it is actually true. It's like convicting an innocent person.
    *   *Example:* A medical test incorrectly indicates that a healthy person has a disease.
*   **Type II Error (False Negative):** This happens when you fail to reject the null hypothesis when it is actually false. It's like letting a guilty person go free.
    *   *Example:* A medical test fails to detect a disease in someone who actually has it.

Question 5: What is the difference between a Z-test and a T-test? Explain when to use each.

Answer:
*   **Z-test:** Used when the population standard deviation is known, or when the sample size is large (typically n > 30). It assumes the data is normally distributed.
*   **T-test:** Used when the population standard deviation is unknown and the sample size is small (typically n ≤ 30). It also assumes the data is normally distributed.

Question 6: Write a Python program to generate a binomial distribution with n=10 and p=0.5, then plot its histogram.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom


n = 10
p = 0.5

binomial_data = binom.rvs(n=n, p=p, size=1000)

plt.hist(binomial_data, bins=range(n + 2), align='left', density=True, alpha=0.7)
plt.xlabel('Number of Successes')
plt.ylabel('Probability')
plt.title('Binomial Distribution (n=10, p=0.5)')
plt.xticks(range(n + 1))
plt.grid(axis='y', alpha=0.75)
plt.show()
```

Question 7: Implement hypothesis testing using Z-statistics for a sample dataset in Python. Show the Python code and interpret the results.

```python
import numpy as np
from statsmodels.stats.weightstats import ztest


sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
               50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
               50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
               50.3, 50.4, 50.0, 49.7, 50.5, 49.9]


mu = 50


z_statistic, p_value = ztest(sample_data, value=mu)


print(f"Z-statistic: {z_statistic}")
print(f"P-value: {p_value}")


alpha = 0.05


if p_value < alpha:
    print(f"Since the p-value ({p_value:.4f}) is less than the significance level ({alpha}), we reject the null hypothesis.")
    print("Conclusion: There is enough evidence to suggest that the population mean is different from 50.")
else:
    print(f"Since the p-value ({p_value:.4f}) is greater than the significance level ({alpha}), we fail to reject the null hypothesis.")
    print("Conclusion: There is not enough evidence to suggest that the population mean is different from 50.")
```

Question 8: Write a Python script to simulate data from a normal distribution and calculate the 95% confidence interval for its mean. Plot the data using Matplotlib.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
from statsmodels.stats.api import DescrStatsW

mu = 75
sigma = 10
sample_size = 100


np.random.seed(42)
simulated_data = np.random.normal(mu, sigma, sample_size)


desc = DescrStatsW(simulated_data)
confidence_interval = desc.tconfint_mean(alpha=0.05)

print(f"Simulated data mean: {np.mean(simulated_data):.4f}")
print(f"95% Confidence Interval for the Mean: ({confidence_interval[0]:.4f}, {confidence_interval[1]:.4f})")

plt.hist(simulated_data, bins=20, density=True, alpha=0.6, color='g')
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, mu, sigma)
plt.plot(x, p, 'k', linewidth=2)
title = "Fit results: mu = %.2f,  std = %.2f" % (mu, sigma)
plt.title(title)
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

Question 9: Write a Python function to calculate the Z-scores from a dataset and visualize the standardized data using a histogram. Explain what the Z-scores represent in terms of standard deviations from the mean.

A Z-score measures how many standard deviations away from the mean a particular data point is.

*   A Z-score of 0 means the data point is exactly at the mean.
*   A positive Z-score means the data point is above the mean.
*   A negative Z-score means the data point is below the mean.

Z-scores allow us to compare data points from different datasets that may have different means and standard deviations, as they standardize the data to a common scale.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import zscore

def calculate_and_plot_zscores(data):
  """
  Calculates Z-scores for a dataset and plots a histogram of the standardized data.

  Args:
    data: A list or NumPy array of numerical data.
  """

  z_scores = zscore(data)


  print("First 10 Z-scores:", z_scores[:10])
  print(f"Mean of Z-scores: {np.mean(z_scores):.4f}")
  print(f"Standard deviation of Z-scores: {np.std(z_scores):.4f}")



  plt.hist(z_scores, bins=20, density=True, alpha=0.7, color='skyblue')
  plt.xlabel('Z-score')
  plt.ylabel('Frequency')
  plt.title('Histogram of Z-scores (Standardized Data)')
  plt.grid(axis='y', alpha=0.75)
  plt.show()

sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
               50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
               50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
               50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

calculate_and_plot_zscores(sample_data)


if 'simulated_data' in locals():
  print("\nUsing simulated data from Question 8:")
  calculate_and_plot_zscores(simulated_data)
else:
  print("\nSimulated data from Question 8 is not available. Please run the code for Question 8 first.")
```
